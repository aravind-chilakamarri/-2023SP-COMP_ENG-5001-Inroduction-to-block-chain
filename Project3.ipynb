```python
# Python 3 code to check 
# available algorithms
  
import hashlib
  
# prints all available algorithms
print ("The available algorithms are : ", end ="")
print (hashlib.algorithms_guaranteed)
```

```python
# Python 3 code to demonstrate
# SHA hash algorithms.
  
# initializing string
str = "Introduction to Blockchain Engineering"
  
# encoding string using encode()
# string is encoded in bytes using UTF-8 (variable length unicode)
# then sending to SHA256
result = hashlib.sha256(str.encode("utf-8"))
 
# printing the equivalent hexadecimal value.
print("The hexadecimal equivalent of SHA256 is : ")
print(result.hexdigest())
  
print ("\n")
  
# Change the first letter from I to J - only 1 bit different
str = "Jntroduction to Blockchain Engineering"

result = hashlib.sha256(str.encode("utf-8"))
  
# printing the equivalent hexadecimal value.
print("The hexadecimal equivalent of SHA256 is : ")
print(result.hexdigest())
```

and/or numbers using a hash algorithm. If a website is hacked for example, cyber criminals don't 
get access to your password. Instead, they just get access to the encrypted “hash” created by 
your password. Say there is a 4-character password (which is very weak). You know the first two 
letters are “AB” and the MD5 hash hexdigest of the entire 4-character string is 
“cd1d14ba40c6a079487fa6952c7da859”. Add another code cell to write a python code to brute-
force the unknown last two characters. Assume each character used in the string is an 
alphanumeric character. Make sure to print each guessed password and its MD5 hash hexdigest. 
Clearly print the 4-character password after finding it and its MD5 hash hexdigest.</b>

```python
import string
import itertools

combinations_generator = itertools.product(string.ascii_uppercase + string.ascii_lowercase + string.digits, 
                                           repeat=2)
combinations = list(map(''.join, combinations_generator))
```

```python
#please dont execute this cell twice. If you wish you execute, please execute the above first and then execute this. Because the combinations will not generate correctly
for each in range(0,len(combinations)):
  combinations[each] = 'AB'+combinations[each]

print(combinations) 
# example combinations: 'AB9D', 'AB9E', 'AB9F', 'AB9G', 'AB9H', 'AB9I', 'AB9J', 'AB9K', 'AB9L',
```

```python
for each in combinations:
  md5hash_result = hashlib.md5(each.encode("utf-8"))
  if md5hash_result.hexdigest() == 'cd1d14ba40c6a079487fa6952c7da859':
    print('The password decoded is ' + each)
    print('The md5 hash for the above password is '+ md5hash_result.hexdigest())
    print('PASSWORD HACKED SUCCESS')
    break
  else:
    print('The password guessed is ' + each)
    print('The md5 hash for the above password guessed is '+ md5hash_result.hexdigest())
    print('PASSWORD HACKED FAILED')
```
